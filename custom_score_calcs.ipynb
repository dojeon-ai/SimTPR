```python
scores_dict_list = [atari_random_scores, atari_human_scores, atari_otr_scores, atari_repr_scores]
norm_scores_dict_list = []
for scores_dict in scores_dict_list:
    norm_scores = {}
    for game, score in scores_dict.items():
        random_score = atari_random_scores[game] 
        human_score = atari_human_scores[game]

        normalized_score = (score - random_score) / (human_score - random_score)
        norm_scores[game] = normalized_score
    norm_scores_dict_list.append(norm_scores)
    
mean_norm_scores = []
med_norm_scores = []
for norm_scores in norm_scores_dict_list:
    mean_norm_scores.append(np.mean(list(norm_scores.values())))
    med_norm_scores.append(np.median(list(norm_scores.values())))
```

```python
col1 = pd.DataFrame.from_dict(scores_dict_list[0], orient='index', columns=['random'])
col2 = pd.DataFrame.from_dict(scores_dict_list[1], orient='index', columns=['human'])
col3 = pd.DataFrame.from_dict(norm_scores_dict_list[2], orient='index', columns=['otr'])
col4 = pd.DataFrame.from_dict(norm_scores_dict_list[3], orient='index', columns=['repr'])

table = pd.concat([col1, col2, col3, col4], axis=1)
table.loc['Mean HNS'] = mean_norm_scores
table.loc['Median HNS'] = med_norm_scores
table.round(3)
```
